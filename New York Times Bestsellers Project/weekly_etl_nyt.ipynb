```python
def book_data_extraction(date, list_names):
    
    """Input
        1. dates: A list of dates or a string of the most recent New York Times bestseller list
        2. list_name: A list of different best seller lists
        
    Output
        A list dataframe with the title, author, publisbher, contributor, 
        published date, list_name, age group and weeks on list of each book """
    
    #book_dicts = []
    #lists = []
    book_data = {'weeks_on_list': [], 'publisher': [], 
                 'title': [], 'author': [], 'contributor': [], 
                 'age_group': [], 'published_date': [], 'list_name': []}
    
    lists_df = pd.read_csv('lists.csv')
    lists =  lists_df['list_name'].str.lower().tolist()
    
    #if type(date) == str:
        #pass
    #elif type(date) == list:
        #pass
    #else:
        #print('The dates given are not correct')
        #return None     

    for list_name in list_names:        
        list_name = list_name.lower()
        
        if list_name not in lists:
            print('One of the lists provided is incorrect. Please provide the correct names')
        else:
            list_name_url = list_name.replace(' ' , '-')
        
            url = f'https://api.nytimes.com/svc/books/v3/lists/{date}/{list_name_url}.json?api-key=ktKhpKGNU9A988ja0j6eDwFu4njxRxC7'
            response = requests.get(url)
            json_data = response.json()

            books_list = json_data['results']['books']
            for book in books_list:
                weeks_on_list = book['weeks_on_list']
                book_data['weeks_on_list'].append(weeks_on_list)

                publisher = book['publisher']
                book_data['publisher'].append(publisher)

                title = book['title']
                book_data['title'].append(title)

                author = book['author']
                book_data['author'].append(author)

                contributor = book['contributor']
                book_data['contributor'].append(contributor)

                age_group = book['age_group']
                book_data['age_group'].append(age_group)

                date_published = json_data['results']['published_date']
                book_data['published_date'].append(date_published)

                list_name = json_data['results']['list_name']
                book_data['list_name'].append(list_name)
                time.sleep(900) 
        
        #book_dicts.append(book_data)
        #df = pd.DataFrame.from_dict(book_data)
        #dataframes.append(df)
        
        
    #dataframes = extract_data(date_chunks, 'hardcover nonfiction')
    #df = pd.concat(dataframes, axis = 'index')
    
    return book_dicts


```

```python
def weekly_book_transformation(df):
    """Input: The dataframe from the weekly_book_extraction function 
    Out: A cleaned version of the data including smaller dataframes 
    that can be loaded into the MySQL tables"""
    
    dataframes = []
    
    df['publisher'] = df.publisher.str.strip()
    df['title'] = df.title.str.capitalize().str.strip()
    df['author'] = df.author.str.strip()
    df['contributor'] = df.contributor.str.replace('by', '').str.strip()

    authors = df['author'].copy()
    dataframes.append(authors)

    publishers = df['publisher'].copy()
    dataframes.append(publishers)
    
    #Creating the dataframe for the 'books table'
    books_temp = df[['title', 'author', 'publisher']].copy()
    books_temp.drop_duplicates(['title', 'author'], keep = 'first', inplace = True)
    books = pd.merge(authors, books_temp, left_on = 'author_name', right_on = 'author', how = 'inner')
    books = pd.merge(books, publishers, left_on = 'publisher', right_on = 'publisher_name', how = 'inner')
    indices = [i for i in range(1, len(books) + 1)]
    books['book_id'] = indices
    
    return dataframes
```
